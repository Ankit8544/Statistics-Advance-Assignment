Q.No-01    What is the Probability density function?

Ans :-

**`The Probability Density Function (PDF)`** is a concept in probability theory and statistics that describes the likelihood of a continuous random variable taking on a particular value within a given range. In simpler terms, it gives you an idea of how the values of a continuous random variable are distributed across its possible range.

For a continuous random variable, the PDF is typically denoted by the function $f(x)$, where $x$ is a specific value within the range of the random variable. The PDF has the following properties:

1. **Non-negative values:** The PDF is always non-negative for all values of $x$: $f(x) \geq 0$.

2. **Area under the curve:** The total area under the PDF curve over the entire range of the random variable is equal to 1. This reflects the fact that the sum of all possible probabilities of the variable's values must be 1.

Mathematically, the probability that the random variable $X$ falls within a specific interval $[a, b]$ is given by the integral of the PDF over that interval:

$$P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx$$

In other words, the probability of $X$ being in the interval $[a, b]$ is the area under the PDF curve between $a$ and $b$.

It's important to note that the PDF only applies to continuous random variables. For discrete random variables, you would use a Probability Mass Function (PMF) to describe the probabilities of specific outcomes.

The concept of the PDF is fundamental in various areas of statistics, including probability distributions (e.g., normal distribution, exponential distribution), hypothesis testing, and estimation. It helps us understand the likelihood of different outcomes and make informed decisions based on the distribution of the data.

**Let's modify the previous example to highlight the interval $[a, b]$ on the PDF graph :**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the normal distribution
mu = 0
sigma = 1

# Generate a range of x values
x = np.linspace(-5, 5, 500)

# Calculate the PDF values for each x using the normal distribution formula
pdf_values = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Define the interval [a, b]
a = -1
b = 1

# Plotting the PDF
plt.plot(x, pdf_values, label=f'PDF (μ={mu}, σ={sigma})')
plt.fill_between(x, pdf_values, where=(x >= a) & (x <= b), color='gray', alpha=0.3, label=f'P({a} ≤ X ≤ {b})')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of a Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

Q.No-02    What are the types of Probability distribution?

Ans :-

Probability distributions are mathematical functions that describe the likelihood of various outcomes in a random experiment. There are several types of probability distributions, each with its own characteristics and applications. Here are some of the most common types:

1. **Discrete Probability Distributions:**
   - **Bernoulli Distribution:** Models a single binary outcome (e.g., success or failure) with a fixed probability.
   - **Binomial Distribution:** Describes the number of successes in a fixed number of independent Bernoulli trials.
   - **Poisson Distribution:** Models the number of events occurring in a fixed interval of time or space, given a certain average rate.

2. **Continuous Probability Distributions:**
   - **Uniform Distribution:** All outcomes within a certain range are equally likely.
   - **Normal (Gaussian) Distribution:** Characterized by its bell-shaped curve and is widely used due to the central limit theorem.
   - **Exponential Distribution:** Describes the time between events in a Poisson process, often used in survival analysis.
   - **Gamma Distribution:** Generalizes the exponential distribution and is used in various applications, including queuing theory.
   - **Beta Distribution:** Models random variables that have values between 0 and 1, often used in Bayesian analysis.
   - **Log-Normal Distribution:** Results from the exponential transformation of a normal distribution, commonly used for modeling positive-valued variables.

3. **Multivariate Distributions:**
   - **Multinomial Distribution:** Generalizes the binomial distribution to more than two categories.
   - **Multivariate Normal Distribution:** Generalization of the normal distribution to multiple dimensions, used in multivariate statistics.
   - **Dirichlet Distribution:** Generalization of the beta distribution to more than two categories, often used in Bayesian analysis for proportions.

4. **Special Distributions:**
   - **Chi-Squared Distribution:** Arises in the context of hypothesis testing and confidence intervals.
   - **Student's t-Distribution:** Used when estimating the mean of a normally distributed population with a small sample size.
   - **F-Distribution:** Often used in analysis of variance (ANOVA) and regression analysis.

Q.No-04    What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

Ans :-

The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials (or experiments), where each trial has only two possible outcomes: success (usually denoted as "S") and failure (usually denoted as "F").

**`The properties of the binomial distribution are as follows :-`**

1. **Fixed Number of Trials (n):** The distribution is defined for a fixed number of trials or experiments, denoted as "n."

2. **Independent Trials:** Each trial is assumed to be independent of the others, meaning the outcome of one trial does not influence the outcome of another trial.

3. **Constant Probability of Success (p):** The probability of success on each trial remains constant and is denoted as "p."

4. **Discrete Outcomes:** The distribution is discrete, meaning it only deals with whole numbers (non-negative integers).

5. **Two Possible Outcomes:** Each trial has only two possible outcomes: success (S) or failure (F).

6. **Binomial Coefficient:** The probability of getting exactly "k" successes out of "n" trials is given by the binomial coefficient (n choose k), which represents the number of ways to choose "k" successes from "n" trials. The formula is:
   
   $P(X = k) = (n choose k) * p^k * (1 - p)^(n - k)$

   Where $(n choose k) = n! / (k! * (n - k)!)$, and "$!$" denotes the factorial of a number.

7. **Probability Mass Function:** The binomial distribution's probability mass function gives the probability of observing a specific number of successes ("$k$") in "$n$" trials.

**`Two examples of events where the binomial distribution can be applied are :-`**

1. **Coin Flips:** Consider flipping a fair coin multiple times. Each flip is a Bernoulli trial, and the outcome is either heads (success) or tails (failure). The goal might be to calculate the probability of getting a certain number of heads in a fixed number of coin flips.

2. **Drug Testing:** In a clinical trial, researchers might be interested in the number of patients out of a sample who respond positively to a new drug. Each patient can be considered a Bernoulli trial with a certain probability of responding to the drug. The researchers might want to calculate the probability of observing a specific number of positive responses in a given number of patients.

In both of these examples, the binomial distribution provides a way to model and analyze the probability of achieving different numbers of successes in a fixed number of independent trials with only two possible outcomes.

Q.N0-06    Write a Python function to calculate the cumulative distribution function of a Poisson distributionwith given mean at a given point.

Ans :-

```python
import math

def poisson_cdf(mean, k):
    if mean < 0 or k < 0:
        raise ValueError("Mean and k must be non-negative values.")
    
    cdf = 0
    for i in range(k + 1):
        cdf += math.exp(-mean) * (mean ** i) / math.factorial(i)
    
    return cdf

# Example usage
mean = float(input("Enter the mean value :- "))
k = int(input("Enter the number of events or occurrences :- "))
cdf_value = poisson_cdf(mean, k)
print("CDF at (k =",k,") :-", cdf_value)
```

Q.No-07    How Binomial distribution different from Poisson distribution?

Ans :-

The Binomial distribution and the Poisson distribution are both probability distributions used in statistics to model the behavior of discrete random variables. They are often used in different contexts and have distinct characteristics:

**`Binomial Distribution :-`**

1. **Nature of Events:** The Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials (experiments with two possible outcomes: success or failure), where the probability of success remains constant across all trials.

2. **Parameters:** It is characterized by two parameters: the number of trials (n) and the probability of success on each trial (p).

3. **Examples:** It's often used to model scenarios like coin flips, where each flip is independent and has two possible outcomes (heads or tails). Another example is the probability of a certain number of defective items in a production line.

4. **Probability Mass Function (PMF):** The probability mass function of the Binomial distribution gives the probability of observing exactly k successes in n trials, and it's given by: 

$P(X = k) = (n choose k) * p^k * (1 - p)^(n - k)$,

where "$n$ choose $k$" is the binomial coefficient.

```python
import scipy.stats as stats

# Parameters for the Binomial distribution
n = 10
p = 0.5

# Calculate probabilities and cumulative probabilities
binomial_pmf = stats.binom.pmf(range(n+1), n, p)
binomial_cdf = stats.binom.cdf(range(n+1), n, p)

# Print probabilities
for k, pmf, cdf in zip(range(n+1), binomial_pmf, binomial_cdf):
    print(f'''
    {k}.    
    P(X = {k}) = {pmf:.4f}
    CDF(X <= {k}) = {cdf:.4f}
    ''')
```

**`Poisson Distribution :-`**

1. **Nature of Events:** The Poisson distribution models the number of events that occur in a fixed interval of time or space, when the events occur randomly and at a constant mean rate.

2. **Parameters:** It is characterized by a single parameter, often denoted as λ (lambda), which represents the average rate of occurrence of the events.

3. **Examples:** The Poisson distribution is commonly used to model the number of phone calls received at a call center in a given hour, the number of accidents occurring at an intersection in a day, or the number of emails received in an hour.

4. **Probability Mass Function (PMF):** The probability mass function of the Poisson distribution gives the probability of observing exactly k events in the given interval, and it's given by: 

$P(X = k) = (e^(-λ) * λ^k) / k!$, 

where "$e$" is the base of the natural logarithm.

```python
import scipy.stats as stats

# Parameter for the Poisson distribution
lambda_ = 5  # Average rate of emails per hour

# Calculate probabilities and cumulative probabilities
poisson_pmf = stats.poisson.pmf(range(11), lambda_)
poisson_cdf = stats.poisson.cdf(range(11), lambda_)

# Print probabilities
for k, pmf, cdf in zip(range(11), poisson_pmf, poisson_cdf):
    print(f'''
    {k}.
    P(X = {k}) = {pmf:.4f}
    CDF(X <= {k}) = {cdf:.4f}
    ''')
```

`In summary`, the key differences between the Binomial and Poisson distributions lie in the nature of events they model, the parameters that characterize them, and the scenarios where they are most applicable. The Binomial distribution deals with a fixed number of trials with two outcomes, while the Poisson distribution deals with a random number of events occurring over a fixed interval with a constant mean rate.

Q.N0-08    Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

Ans :-

```python
import numpy as np

# Parameters
mean = 5
sample_size = 1000

# Generate random sample
random_sample = np.random.poisson(mean, sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample, ddof=1)  # Use ddof=1 for sample variance

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
```

Q.No-09    How mean and variance are related in Binomial distribution and Poisson distribution?

Ans :-

The mean and variance of a probability distribution provide important insights into its shape and characteristics. Let's discuss how mean and variance are related in the context of the Binomial distribution and the Poisson distribution.

**`Binomial Distribution :-`**


The Binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials, where each trial has a fixed probability of success, denoted as "$p$".

For a Binomial distribution, the mean ($μ$) is given by:
$μ = n * p$,

where "$n$" is the number of trials and "$p$" is the probability of success in each trial.

The variance ($σ^2$) of a Binomial distribution is given by:

$σ^2 = n * p * (1 - p)$,

where, again, "$n$" is the number of trials and "$p$" is the probability of success in each trial.

```python
import numpy as np
from scipy.stats import binom, poisson

# Binomial Distribution
n = 10  # Number of trials
p = 0.3  # Probability of success
mean_binomial = n * p
variance_binomial = n * p * (1 - p)

print("Binomial Distribution :-")
print(f"Mean: {mean_binomial}")
print(f"Variance: {variance_binomial}")
```

**`Poisson Distribution :-`**

The Poisson distribution models the number of events occurring in a fixed interval of time or space, given a known average rate of occurrence, denoted as "$λ$" (lambda).

For a Poisson distribution, both the mean ($μ$) and the variance ($σ^2$) are given by the average rate of occurrence, "$λ$":

$μ = λ$,

$σ^2 = λ$.

```python
import numpy as np
from scipy.stats import binom, poisson

# Poisson Distribution
lambda_poisson = 5  # Average rate of occurrence
mean_poisson = lambda_poisson
variance_poisson = lambda_poisson

print("Poisson Distribution:")
print(f"Mean: {mean_poisson}")
print(f"Variance: {variance_poisson}")
```

In summary, for both the Binomial and Poisson distributions:

1. In the Binomial distribution, the variance increases as the probability of success "p" moves away from 0.5 (becomes more skewed), and it reaches its maximum value when "$p$" is 0.5. When "$p$" is close to 0 or 1, the distribution becomes more concentrated and the variance decreases.

2. In the Poisson distribution, both the mean and the variance are equal and are determined solely by the average rate of occurrence "λ." As "λ" increases, the distribution becomes more spread out.

It's important to note that these relationships provide insights into the shape of these distributions, but individual instances may still vary due to the inherent randomness of the processes being modeled.

Q.No-10    In normal distribution with respect to mean position, where does the least frequent data appear?

Ans :-

In a normal distribution (also known as a Gaussian distribution), the data is symmetrically distributed around the mean. This means that the least frequent data points appear in the tails of the distribution, far away from the mean.

In a standard normal distribution (a specific type of normal distribution with a mean of 0 and a standard deviation of 1), approximately 68% of the data falls within one standard deviation of the mean, about 95% falls within two standard deviations, and roughly 99.7% falls within three standard deviations. This leaves only a very small percentage (0.3%) of data points in the tails, which are the least frequent data points.

    To be more specific, in a standard normal distribution:
- Data points that are more than 3 standard deviations below the mean or more than 3 standard deviations above the mean would be considered the least frequent data points.
- Data points between 2 and 3 standard deviations away from the mean are less frequent but still more common than the ones beyond 3 standard deviations.
- Data points between 1 and 2 standard deviations away from the mean are more frequent, encompassing a larger portion of the distribution.

In summary, the least frequent data points in a normal distribution appear in the tails, which are the regions that are farthest from the mean.
